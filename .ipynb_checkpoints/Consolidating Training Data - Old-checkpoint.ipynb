```python
def coordinates_left(value):
    if value > 0:
        return 1
    elif value < 0:
        return 0
    else:
        return 0
    
def coordinates_right(value):
    if value > 0:
        return 0
    elif value < 0:
        return 1
    else:
        return 0
    
def coordinates_above(value):
    if value > 0:
        return 1
    elif value < 0:
        return 0
    else:
        return 0
    
def coordinates_below(value):
    if value > 0:
        return 0
    elif value < 0:
        return 1
    else:
        return 0
```

```python
def clear(value):
    if abs(value) < 1:
        return 1
    else:
        return 0
```

```python
left['right_of'] = left['p1x'] - left['p2x'] 
left['left_of'] = left['p1x'] - left['p2x'] 
left['above'] = left['p1y'] - left['p2y'] 
left['below'] = left['p1y'] - left['p2y'] 

left['x_clear'] = left['right_of'].apply(lambda x: clear(x))
left['y_clear'] = left['above'].apply(lambda x: clear(x))

left['left_of'] = left['left_of'].apply(lambda x: coordinates_left(x))
left['right_of'] = left['right_of'].apply(lambda x: coordinates_right(x))
left['below'] = left['below'].apply(lambda x: coordinates_below(x))
left['above'] = left['above'].apply(lambda x: coordinates_above(x))

```

```python
# assign awards to the proper location

p_reward = list(mframe[mframe['reward']>0].index)

for reward_index in p_reward:
    try:
        index_back = 9

        sub = mframe[reward_index - index_back:reward_index + 1]
        reward = sub.ix[reward_index]['reward']

        cd1_max = max(sub['cd1'])
        cd2_max = max(sub['cd2'])
        
        if cd1_max >= cd2_max:
            action = sub[sub['action'] == 6]
            action = action[action['cd1'] != 1]
            action = action[action['reward'] == 0]
            action_index = list(action.index)[-1]
            mframe.ix[action_index, 'reward'] = reward
            mframe.ix[reward_index, 'reward'] = 0
        else:
            action = sub[sub['action'] == 7]
            action = action[action['cd2'] != 1]
            action = action[action['cd2'] != 2]
            action = action[action['reward'] == 0]
            action_index = list(action.index)[-1]
            mframe.ix[action_index, 'reward'] = reward
            mframe.ix[reward_index, 'reward'] = 0
       
    except:
        pass
```

```python
# replace negative actions

negative_actions = mframe[mframe['reward']<0]
negative_action_7 = list(negative_actions[negative_actions['action'] == 7].index)
negative_action_6 = list(negative_actions[negative_actions['action'] == 6].index)

negative_actions = []
negative_actions.extend(negative_action_7)
negative_actions.extend(negative_action_6)
len(negative_actions)
```

```python
for index in negative_actions:
    action = 4
    mframe.ix[int(index), 'action'] == action
```

```python
#remove negative rewards
negative_rewards = list(mframe[mframe['reward'] < 0].index)
for index in negative_rewards:
    mframe.ix[int(index), 'reward'] = 0
```
